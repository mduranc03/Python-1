```python
#La siguiente es la estructura que podemos ver en el link
'https://www.football-data.co.uk/mmz4281/'+'2425'+'/'+'E0'+'.csv'
```

```python
#Ya se que la raiz es la misma asi que la guardo en una variable que se llame ruta
ruta='https://www.football-data.co.uk/mmz4281/'
```

```python
#Ahora voy a crear la lista con el E1,E2,E3,Ec
campeonatos=['E0', 'E1', 'E2', 'E3', 'EC']
enlaces=[]

#Para que esto funcione debe recorrer la lista con un loop, a esta practica se le conoce como
#Link dinamico, una vez recorra la lista debe remplazar el valor para que el enlace funcione

for campeonato in campeonatos:
  df_links=pd.read_csv(ruta + '2425' + '/'+ campeonato + '.csv')
  enlaces.append(df_links)
```
